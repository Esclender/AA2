# Pandas Data Analysis Notebook

This notebook demonstrates the usage of the pandas library for data analysis.

```python
# Import pandas library
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv"
df = pd.read_csv(url)

# show the first 5 rows of the dataset
print(df.head())

```

```python
# show the last 5 rows of the dataset
print(df.tail())
```

```python
# show data types
print(df.dtypes)
```

```python
# Indentify null values
print(df.isnull().sum())
```

```python
# Describe critical points in the dataset
print(df.describe())
```

```python
import matplotlib.pyplot as plt

df.boxplot(column=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'], figsize=(10, 6))
plt.title('Boxplot to Identify Extreme Values')
plt.ylabel('Values')
plt.show()
```
